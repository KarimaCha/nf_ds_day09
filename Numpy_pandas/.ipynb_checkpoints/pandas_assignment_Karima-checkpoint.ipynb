3) I've decided that this month I want to stay away from wines with relatively high alcohol content. To do that, I'm going to avoid any wines that have a greater alcohol content than the mean alcohol content, and you're going to help me do this. To achieve this, let's do the following: 

  * Find the mean alcohol content, separately, for reds and whites.  
  * Create a `Series` that holds whether each row in each `DataFrame` (the reds, and whites) has a higher alcohol content than the mean. 
  * Merge this `Series` onto the `DataFrame`. I can imagine doing this with either a `.join()` or using `pd.concat()`. For practice, do it with both. Note: merges with `Series` work the same way that they work with `DataFrames`.  
  * Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.  
   
   


```python
# Find the mean alcohol content, separately, for reds and whites.
alc_mean_white = white_df['alcohol'].mean()
alc_mean_red = red_df['alcohol'].mean()
alc_mean_white, alc_mean_red
```

```python
# Create a Series that holds whether each row in each DataFrame 
#   (the reds, and whites) has a higher alcohol content than the mean.
alc_high_white = white_df['alcohol'] > alc_mean_white #['alcohol']
alc_high_red = red_df['alcohol'] > alc_mean_red #['alcohol']
alc_high_white.name = 'alc_above_mean'
alc_high_red.name = 'alc_above_mean'

print(alc_high_white.head())
print(alc_high_red.head())
```

```python
# Merge this Series onto the DataFrame. I can imagine doing this with either 
#   a .join() or using pd.concat(). For practice, do it with both. 
#   Note: merges with Series work the same way that they work with DataFrames.

# a) with join
white_df_join = white_df.join(alc_high_white)
red_df_join = red_df.join(alc_high_red)
red_df_join.head()
```

```python
# b) with concat
white_df_concat = pd.concat([white_df, alc_high_white], axis = 1)
red_df_concat = pd.concat([red_df, alc_high_red], axis = 1)
red_df_concat.head()
```

```python
# Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.

white_df_high_alc = white_df_concat[white_df_concat['alc_above_mean']]
red_df_high_alc = red_df_concat[red_df_concat['alc_above_mean']]
white_df_high_alc.head()
```

```python
red_df_high_alc.head()
```

```python
# Note: Option 2 without constructing an extra df inbetween:
white_df_high_alc2 = white_df[white_df['alcohol'] > white_df['alcohol'].mean()]
red_df_high_alc2 = red_df[red_df['alcohol'] > red_df['alcohol'].mean()]

red_df_high_alc2.head()
```

```python
# Check if option 1 and option 2 return the same results
all(red_df_high_alc.iloc[:,:-1] == red_df_high_alc2)
all(white_df_high_alc.iloc[:,:-1] == white_df_high_alc2)
```

4) Let's say that I want to get started on cutting back next month. This time, though, I want to focus on staying away from those wines with a high acidity. Specifically, I want to stay away from those wines that are in the highest bin of fixed acidity (highest bin out of 5). You're now going to help me do this. To achieve this, let's do the following: 

 * Separate the rows in each `DataFrame` into 5 equal width bins based off their fixed acidity. 
 * Merge the resulting `Series` holding these 5 bins onto the original `DataFrame`. I can imagine also doing this with either `.join()` or using `pd.concat()`. Try doing it with both for practice. 
 * Return back to me all those rows that are **not in ** the top bin in terms of fixed acidity. 



```python
# Separate the rows in each DataFrame into 5 equal width bins based off their fixed acidity.
import numpy as np
red_df['fixed acidity'].min(), red_df['fixed acidity'].max()
fixed_acidity_bins_red = np.linspace(red_df['fixed acidity'].min(), red_df['fixed acidity'].max(), 6)
fixed_acidity_series_red = pd.cut(red_df['fixed acidity'], bins=fixed_acidity_bins_red, 
                              labels=range(1,6))   # fixed_acidity_bins_red[:-1]
fixed_acidity_series_red.name = 'fa_bin'

white_df['fixed acidity'].min(), white_df['fixed acidity'].max()
fixed_acidity_bins_white = np.linspace(white_df['fixed acidity'].min(), white_df['fixed acidity'].max(), 6)
fixed_acidity_series_white = pd.cut(white_df['fixed acidity'], bins=fixed_acidity_bins_white, 
                              labels=range(1,6))   # fixed_acidity_bins_white[:-1])
fixed_acidity_series_white.name = 'fa_bin'

fixed_acidity_series_red.head()
fixed_acidity_series_white.head()
```

```python
# Merge the resulting Series holding these 5 bins onto the original DataFrame. 
# I can imagine also doing this with either .join() or using pd.concat(). 
# Try doing it with both for practice.

# a) with concat
red_df_concat2 = pd.concat([red_df, fixed_acidity_series_red], axis=1)
red_df_concat2.head()

white_df_concat2 = pd.concat([white_df, fixed_acidity_series_white], axis=1)
white_df_concat2.head()
```

```python
# b) with join
red_df_join2 = red_df.join(fixed_acidity_series_red)
red_df_join2.head()

white_df_join2 = white_df.join(fixed_acidity_series_white)
white_df_join2.head()
```

```python
# Return back to me all those rows that are NOT in the top bin in terms of fixed acidity
red_df_low_acidity = red_df_join2[red_df_join2['fa_bin']!=5]
white_df_low_acidity = white_df_join2[white_df_join2['fa_bin']!=5]

white_df_low_acidity.head()
```

5) Let's say that I now want to know how much my decision to avoid those wines with higher `alcohol` content is going to limit the `quality` of wines that I can drink. To figure this out, I want to know a couple of things: 

 * The average `alcohol` content for those reds above the mean `alcohol` level, by quality.
 * The average `alcohol` content for those whites above the mean `alcohol` level, by quality. 
 
 Use a `pivot table` to solve this. 
 


```python
pivot_red_alc = pd.pivot_table(data = red_df_concat, values='alcohol', index='quality', 
                           columns='alc_above_mean')  #, aggfunc=np.mean)
pivot_red_alc[True].reset_index()
```

```python
pivot_white_alc = pd.pivot_table(data = white_df_concat, values='alcohol', index='quality', 
                             columns='alc_above_mean')  #, aggfunc=np.mean)
pivot_white_alc
```
